```python
n=np.arange(-10,11)
n
```

```python
def calculate_xn(i):
    if 0 <= i <= 10:
        return i*i + i
    elif -10 <= i < 0:
        return -i*i + i
    else:
        return 0

# Calculate corresponding y values
xn = np.array([calculate_xn(i) for i in n])
xn
```
