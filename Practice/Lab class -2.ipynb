# Unite step Signal

```python
# init step signal
# unit impulse signal
# unit ramp signal
#n**2 for 0<N<5
#N FOR N>=5
#-N for n<0
```

```python
def unitstep(n):
    y=[]
    for i in n:
        if i>=0:
            y.append(1)
        else: y.append(0)
    return y

n=np.arange(-10,10)
y=unitstep(n)
plt.stem(n,y)
#plt.show
```

# Unit Ramp

```python
def unitramp(n):
    y=[]
    for i in n:
        if i>=0:
            y.append(i)
        else: y.append(0)
    return y
n=np.arange(-10,10)
y=unitramp(n)
plt.stem(n,y)
plt.show
```

# Custom Function

```python
def customeFun(n):
    y=[]
    for temp in n:
        if temp>=0 and temp<5:
            y.append(temp*temp)
        elif temp>=5: y.append(temp)
        else: y.append(-temp)
    return y
n=np.arange(-10,10)
y=customeFun(n)
plt.stem(n,y)
plt.show
```
